```python
import math

def sin_cos_series(x, n):
    """
    Compute the nth order Taylor expansion for sin(x) and cos(x) around 0.
    """
    # Initialize the series counters
    sin_series = 0
    cos_series = 0

    # Loop over the terms in the series
    for i in range(n):
        # Compute the ith term in the sin series
        sin_term = (-1)**i * x**(2*i+1) / math.factorial(2*i+1)
        sin_series += sin_term

        # Compute the ith term in the cos series
        cos_term = (-1)**i * x**(2*i) / math.factorial(2*i)
        cos_series += cos_term

    return sin_series, cos_series

def sin_cos_product_series(x, n):
    """
    Compute the nth order Taylor expansion for sin(x)cos(x) around 0.
    """
    # Initialize the series counter
    sin_cos_series = 0

    # Loop over the terms in the series
    for i in range(n):
        # Compute the ith term in the series
        term = (-1)**i * x**(2*i+1) / math.factorial(2*i+1)
        sin_cos_series += term

    return sin_cos_series


# Compute Taylor expansions for sin(x), cos(x), and sin(x)*cos(x)
x = math.pi/2
n = 4

sin_expansion, cos_expansion = sin_cos_series(x, n)
product_expansion = sin_cos_product_series(x, n)

# Print the results
print(f"sin(x) = {sin_expansion}")
print(f"cos(x) = {cos_expansion}")
print(f"sin(x)*cos(x) = {product_expansion}")

# Compare errors
sin_cos_error = abs(math.sin(x)*math.cos(x) - product_expansion)
sin_cos_separate_error = abs(math.sin(x) - sin_expansion) * abs(math.cos(x) - cos_expansion)

if sin_cos_error < sin_cos_separate_error:
    print("The Taylor expansion for the product first then plugging in x produces less error.")
else:
    print("Computing the Taylor expansion for sin and cos separately then multiplying the result together produces less error.")


# Compute the approximation and truncation error bound for cos(pi/4)
def cos_series(x, n):
    """
    Compute the nth order Taylor expansion for cos(x) around 0.
    """
    # Initialize the series counter
    cos_series = 0

    # Loop over the terms in the series
    for i in range(n):
        # Compute the ith term in the cos series
        cos_term = (-1)**i * x**(2*i) / math.factorial(2*i)
        cos_series += cos_term

    return cos_series

def main():
    x = math.pi/4
    n = 4

    approx = cos_series(x, n)
    print("The approximation is:", approx)

    error = abs(math.cos(x) - approx)
    print("The error is:", error)

    truncation_error = x**5/120
    print("The truncation error bound is:", truncation_error)

main()
```
